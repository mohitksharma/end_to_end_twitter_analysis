```python
import configparser
config = configparser.RawConfigParser()
config.read('twitter.properties')
print(config.sections());
```

```python
accesstoken=config.get('twitter', 'access_token')
accesstokensecret=config.get('twitter', 'access_token_secret')
apikey=config.get('twitter', 'api_key')
apisecretkey=config.get('twitter', 'api_secret_key')
```

```python
import os
import tweepy as tw
```

```python
auth = tw.OAuthHandler(apikey, apisecretkey)
auth.set_access_token(accesstoken, accesstokensecret)
api = tw.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
search_words = "#coronavirus"
date_since = "2020-05-11"
```

```python
tweets = tw.Cursor(api.search,
                   q=search_words,
                   lang="en", 
                   tweet_mode = 'extended',
                   truncated = False,
                   since=date_since).items(5)

#for tweet in tweets:
    #print(tweet)
```

```python
tweets = tw.Cursor(api.search,
                   q=search_words,
                   lang="en", 
                   tweet_mode = 'extended',
                   truncated = False,
                   since=date_since).items(10000)

tweet_details = [[tweet.geo, tweet.full_text,tweet.user.screen_name, tweet.user.location, tweet.entities['hashtags'], tweet.source, tweet.user.followers_count, tweet.user.friends_count, tweet.created_at, tweet.user.verified ]for tweet in tweets]
```

```python
tweet_df = pd.DataFrame(tweet_details, columns=['geo', 'text', 'name', 'location', 'hashtags', 'source', 'followers_count', 'friends_count', 'created_at', 'verification_status'])
```

```python
for row in range(0, len(tweet_df['hashtags'])):
    text = []
    for data in tweet_df['hashtags'][row]:
        text.append(data['text'])
    tweet_df['hashtags'][row] = text
```

```python
def clean_text(text):
    text = re.sub("RT @[\w]*:","", text)
    text = re.sub("https://[a-zA-Z0-9].*", "", text)
    text = re.sub("@[\w]*","", text)
    text = re.sub("\n","",text)
    return text
```

```python
tweet_df['text'] = tweet_df['text'].apply(clean_text)
```

```python
import spacy 
  
nlp = spacy.load('en_core_web_sm')
```

```python
tweet_df['text'].apply(lambda x: [print("\tText : {}, Entity : {}".format(ent.text, ent.label_)) if (not ent.text.startswith('#')) else ""  for ent in nlp(x).ents])
```

```python
tweet_df['entities']=tweet_df['text'].apply(lambda x: [(ent.text, ent.label_) if (not ent.text.startswith('#')) else "" for ent in nlp(x).ents])
```
